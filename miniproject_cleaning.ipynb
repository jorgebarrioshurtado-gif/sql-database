```python
df = df.drop(columns = ['DailyRate', 'Education', 'EducationField', 'EmployeeCount', 'Gender', 'HourlyRate',
       'JobLevel', 'MaritalStatus', 'MonthlyRate', 'Over18', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TotalWorkingYears'])
```

```python
df.shape
```

We started with 35 columns and left only 19. We saved the new dataset in a csv file. This dataset is going to be the dataset that we'll use in the project.
